   <h2 id="number">Python Data Types</h2>

<p>Every value in Python has a datatype. Since everything is an object in Python programming, data types are actually classes and variables are instance (object) of these classes.</p>

<p>There are various data types in Python. Some of the important types are listed below.</p>

<hr><h2 id="number">Python Numbers</h2>

<p>Integers, floating point numbers and complex numbers fall under <a href="https://www.programiz.com/python-programming/numbers">Python numbers</a> category. They are defined as <code>int</code>, <code>float</code> and <code>complex</code> classes in Python.</p>

<p>We can use the <code>type()</code> function to know which class a variable or a value belongs to. Similarly, the <code>isinstance()</code> function is used to check if an object belongs to a particular class.</p>

<pre class="python-exec">
<code>a = 5
print(a, "is of type", type(a))

a = 2.0
print(a, "is of type", type(a))

a = 1+2j
print(a, "is complex number?", isinstance(1+2j,complex))
</code></pre>

<p><strong>Output</strong></p>

<pre>
<samp>5 is of type &lt;class 'int'&gt;
2.0 is of type &lt;class 'float'&gt;
(1+2j) is complex number? True
</samp></pre>

<p>Integers can be of any length, it is only limited by the memory available.</p>

<p>A floating-point number is accurate up to 15 decimal places. Integer and floating points are separated by decimal points. <samp>1</samp> is an integer, <samp>1.0</samp> is a floating-point number.</p>

<p>Complex numbers are written in the form, <code>x + yj</code>, where <var>x</var> is the real part and <var>y</var> is the imaginary part. Here are some examples.</p>

<pre>
<code>&gt;&gt;&gt; a = 1234567890123456789
&gt;&gt;&gt; a
1234567890123456789
&gt;&gt;&gt; b = 0.1234567890123456789
&gt;&gt;&gt; b
0.12345678901234568
&gt;&gt;&gt; c = 1+2j
&gt;&gt;&gt; c
(1+2j)
</code></pre>

<p>Notice that the <code>float</code> variable <var>b</var> got truncated.</p>

<hr><h2 id="list">Python List</h2>

<p><a href="https://www.programiz.com/python-programming/list">List</a> is an ordered sequence of items. It is one of the most used datatype in Python and is very flexible. All the items in a list do not need to be of the same type.</p>

<p>Declaring a list is pretty straight forward. Items separated by commas are enclosed within brackets <code>[ ]</code>.</p>

<pre>
<code>a = [1, 2.2, 'python']
</code></pre>

<p>We can use the slicing operator <code>[ ]</code> to extract an item or a range of items from a list. The index starts from 0 in Python.</p>

<pre class="python-exec">
<code>a = [5,10,15,20,25,30,35,40]

# a[2] = 15
print("a[2] = ", a[2])

# a[0:3] = [5, 10, 15]
print("a[0:3] = ", a[0:3])

# a[5:] = [30, 35, 40]
print("a[5:] = ", a[5:])
</code></pre>

<p><strong>Output</strong></p>

<pre>
<samp>a[2] =  15
a[0:3] =  [5, 10, 15]
a[5:] =  [30, 35, 40]
</samp></pre>

<p>Lists are mutable, meaning, the value of elements of a list can be altered.</p>

<pre class="python-exec">
<code>a = [1, 2, 3]
a[2] = 4
print(a)
</code></pre>

</div><div class="clearfix"></div><p><strong>Output</strong></p>

<pre>
<samp>[1, 2, 4]
</samp></pre>

<hr><h2 id="tuple">Python Tuple</h2>

<p><a href="https://www.programiz.com/python-programming/tuple">Tuple</a> is an ordered sequence of items same as a list. The only difference is that tuples are immutable. Tuples once created cannot be modified.</p>

<p>Tuples are used to write-protect data and are usually faster than lists as they cannot change dynamically.</p>

<p>It is defined within parentheses <code>()</code> where items are separated by commas.</p>

<pre>
<code>t = (5,'program', 1+3j)
</code></pre>

<p>We can use the slicing operator <code>[]</code> to extract items but we cannot change its value.</p>

<pre class="python-exec">
<code>t = (5,'program', 1+3j)

# t[1] = 'program'
print("t[1] = ", t[1])

# t[0:3] = (5, 'program', (1+3j))
print("t[0:3] = ", t[0:3])

# Generates error
# Tuples are immutable
t[0] = 10
</code></pre>

<p><strong>Output</strong></p>

<pre>
<samp>t[1] =  program
t[0:3] =  (5, 'program', (1+3j))
Traceback (most recent call last):
File "test.py", line 11, in &lt;module&gt;
t[0] = 10
TypeError: 'tuple' object does not support item assignment
</samp></pre>

<hr><h2 id="string">Python Strings</h2>

<p><a href="https://www.programiz.com/python-programming/string">String</a> is sequence of Unicode characters. We can use single quotes or double quotes to represent strings. Multi-line strings can be denoted using triple quotes, <code>'''</code> or <code>"""</code>.</p>

<pre class="python-exec">
<code>s = "This is a string"
print(s)
s = '''A multiline
string'''
print(s)
</code></pre>

<p><strong>Output</strong></p>

<pre>
<samp>This is a string
A multiline
string
</samp></pre>

<p>Just like a list and tuple, the slicing operator <code>[ ]</code> can be used with strings. Strings, however, are immutable.</p>

<pre class="python-exec">
<code>s = 'Hello world!'

# s[4] = 'o'
print("s[4] = ", s[4])

# s[6:11] = 'world'
print("s[6:11] = ", s[6:11])

# Generates error
# Strings are immutable in Python
s[5] ='d'
</code></pre>

<p><strong>Output</strong></p>

<pre>
<samp>s[4] =  o
s[6:11] =  world
Traceback (most recent call last):
File "&lt;string&gt;", line 11, in &lt;module&gt;
TypeError: 'str' object does not support item assignment
</samp></pre>

<hr><h2 id="set">Python Set</h2>

<p><a href="https://www.programiz.com/python-programming/set">Set</a> is an unordered collection of unique items. Set is defined by values separated by comma inside braces <code>{ }</code>. Items in a set are not ordered.</p>

<pre class="python-exec">
<code>a = {5,2,3,1,4}

# printing set variable
print("a = ", a)

# data type of variable a
print(type(a))
</code></pre>

<p><strong>Output</strong></p>

<pre>
<samp>a =  {1, 2, 3, 4, 5}
&lt;class 'set'&gt;
</samp></pre>

<p>We can perform set operations like union, intersection on two sets. Sets have unique values. They eliminate duplicates.</p>

<pre class="python-exec">
<code>a = {1,2,2,3,3,3}
print(a)
</code></pre>

<p><strong>Output</strong></p>

<pre>
<samp>{1, 2, 3}
</samp></pre>

<p>Since, set are unordered collection, indexing has no meaning. Hence, the slicing operator <code>[]</code> does not work.</p>

<pre>
<code>&gt;&gt;&gt; a = {1,2,3}
&gt;&gt;&gt; a[1]
Traceback (most recent call last):
File "&lt;string&gt;", line 301, in runcode
File "&lt;interactive input&gt;", line 1, in &lt;module&gt;
TypeError: 'set' object does not support indexing
</code></pre>

<hr><h2 id="dic">Python Dictionary</h2>

<p><a href="https://www.programiz.com/python-programming/dictionary">Dictionary</a> is an unordered collection of key-value pairs.</p>

<p>It is generally used when we have a huge amount of data. Dictionaries are optimized for retrieving data. We must know the key to retrieve the value.</p>

<p>In Python, dictionaries are defined within braces <code>{}</code> with each item being a pair in the form <code>key:value</code>. Key and value can be of any type.</p>

<pre>
<code>&gt;&gt;&gt; d = {1:'value','key':2}
&gt;&gt;&gt; type(d)
&lt;class 'dict'&gt;
</code></pre>

<p>We use key to retrieve the respective value. But not the other way around.</p>

<pre class="python-exec">
<code>d = {1:'value','key':2}
print(type(d))

print("d[1] = ", d[1])

print("d['key'] = ", d['key'])

# Generates error
print("d[2] = ", d[2])
</code></pre>

<p><strong>Output</strong></p>

<pre>
<samp>&lt;class 'dict'&gt;
d[1] =  value
d['key'] =  2
Traceback (most recent call last):
File "&lt;string&gt;", line 9, in &lt;module&gt;
KeyError: 2
</samp></pre>

<hr><h2 id="con">Conversion between data types</h2>

<p>We can convert between different data types by using different type conversion functions like <code>int()</code>, <code>float()</code>, <code>str()</code>, etc.</p>

<pre>
<code>&gt;&gt;&gt; float(5)
5.0
</code></pre>

<p>Conversion from float to int will truncate the value (make it closer to zero).</p>

<pre>
<code>&gt;&gt;&gt; int(10.6)
10
&gt;&gt;&gt; int(-10.6)
-10
</code></pre>

<p>Conversion to and from string must contain compatible values.</p>

<pre>
<code>&gt;&gt;&gt; float('2.5')
2.5
&gt;&gt;&gt; str(25)
'25'
&gt;&gt;&gt; int('1p')
Traceback (most recent call last):
File "&lt;string&gt;", line 301, in runcode
File "&lt;interactive input&gt;", line 1, in &lt;module&gt;
ValueError: invalid literal for int() with base 10: '1p'
</code></pre>

<p>We can even convert one sequence to another.</p>

<pre>
<code>&gt;&gt;&gt; set([1,2,3])
{1, 2, 3}
&gt;&gt;&gt; tuple({5,6,7})
(5, 6, 7)
&gt;&gt;&gt; list('hello')
['h', 'e', 'l', 'l', 'o']
</code></pre>

<p>To convert to dictionary, each element must be a pair:</p>

<pre>
<code>&gt;&gt;&gt; dict([[1,2],[3,4]])
{1: 2, 3: 4}
&gt;&gt;&gt; dict([(3,26),(4,44)])
{3: 26, 4: 44}
</code></pre>
